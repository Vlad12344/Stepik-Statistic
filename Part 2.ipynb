```python
# Step 3
# array = np.loadtxt(".../.../*.txt", comments="#", delimiter=",", unpack=False)
# print('Mean:', np.mean(array))

# Step 5
"_______2/6*(x+y+z+t+k+s)______"

# Step 6
array = [17.4, 19.9, 19.9, 19.0, 20.7, 18.2, 18.6,
         20.3, 21.7, 20.8, 16.7, 20.3, 22.5, 18.3,
         21.1, 16.3, 22.0, 19.5, 18.1, 21.1, 20.0,
         21.6, 21.5, 21.1, 23.3, 23.0, 17.8, 19.5, 21.6, 22.9]
print('Dispersiya:', sum(1/len(array) * (array - np.mean(array))**2))

# Step 7
"_______sqrt(1/5*(x^2+y^2+z^2+t^2+s^2-5*a^2))_______"

# Step 10
sample = [2,  1,  0,  0,  1]
lmd = 3

        # Puasson
def ln_likelihood_fnc(sample, lmd):
    fn = 0
    for k in sample:
        fn += np.log(np.exp(-lmd) * math.pow(lmd, k) / math.factorial(k))
    
    return fn

print("ln likelihood fnc:", ln_likelihood_fnc(sample, lmd))

# Step 11
sample = np.loadtxt(os.path.join(root, '25_11.txt'), comments="#", delimiter=",", unpack=False)

def max_likelihood_fnc(sample):
    fn = list(map(lambda x: 1/x, sample))

    return sample[np.argmax(fn)]

print('Max likelihood:', max_likelihood_fnc(sample))

# Step 12
""""
    >–ù–∏ –æ–¥–Ω–æ –∏–∑ —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–π –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –≤–µ—Ä–Ω—ã–º
     –û—Ü–µ–Ω–∫–∏, –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –º–µ—Ç–æ–¥–æ–º –º–æ–º–µ–Ω—Ç–æ–≤, —è–≤–ª—è—é—Ç—Å—è –Ω–µ—Å–º–µ—â–µ–Ω–Ω—ã–º–∏
     –û—Ü–µ–Ω–∫–∏, –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –º–µ—Ç–æ–¥–æ–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è, —è–≤–ª—è—é—Ç—Å—è –Ω–µ—Å–º–µ—â–µ–Ω–Ω—ã–º–∏
     –í—Å–µ –º–µ—Ç–æ–¥—ã –¥–∞—é—Ç –Ω–µ—Å–º–µ—â–µ–Ω–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏
     –û—Ü–µ–Ω–∫–∏, –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –º–µ—Ç–æ–¥–æ–º –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–∏, —è–≤–ª—è—é—Ç—Å—è –Ω–µ—Å–º–µ—â–µ–Ω–Ω—ã–º–∏
"""

# Step 13
"_______6*1/(x+y+z+t+k+s)_______"

# Step 15
"""
    –í–µ—Ä–Ω–æ –ª–∏ –¥–∞–Ω–Ω–æ–µ —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ?

    –ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ ùúÉÃÇ  —è–≤–ª—è–µ—Ç—Å—è –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∏ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ–π –æ—Ü–µ–Ω–∫–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ ùúÉ 
    –≤ –∫–ª–∞—Å—Å–µ –Ω–µ—Å–º–µ—â–µ–Ω–Ω—ã—Ö –æ—Ü–µ–Ω–æ–∫, –µ—Å–ª–∏ limùëõ‚Üí‚àûùê∑ùúÉÃÇ 1/[ùëõùêº(ùúÉ)]=1, –≥–¥–µ ùêº(ùúÉ) - –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –§–∏—à–µ—Ä–∞ –æ –ø–∞—Ä–∞–º–µ—Ç—Ä–µ  ùúÉ

    > –î–∞
"""

# Step 16
```
