### 3.2 Доверительные интервалы

```python
# Step 7
"""
    На основе продолжительных наблюдений за весом Х г. пакетов с орешками, 
    заполняемых автоматически, установлено, что среднее квадратическое отклонение 
    веса пакетов равно 10г. Средний вес 100 наудачу выбранных пакетов оказался равным 249 г. 
    Укажите доверительный интервал для среднего веса пакетов, построенный с доверительной вероятностью 0.9. 
    Предполагается, что распределение веса пакетиков подчиняется нормальному закону распределения.

    Формат ответа: границы интервала через запятую с пробелом, вводятся с точностью 3 знака после запятой. 
    Сначала левая граница, потом правая
"""
# Для доверительной вероятности 0.9 интервал равен 1.645
X_, sigma, n, u = 249, 10, 100, 1.645

left_bound = X_ - u * sigma/math.sqrt(n)
right_bound = X_ + u * sigma/math.sqrt(n)

print("Step 7\n", "Bounds:", "%.3f" % left_bound, '< a <', "%.3f" % right_bound)

# Step 9
"""
    Передположим, что диаметр шариков, изготавливаемых станком-автоматом, 
    подчиняется нормальному закону распределения. В  выборке  из 16  шариков средний  
    диаметр  оказался  равным  10.3 мм,   а  несмещенная оценка дисперсии - 1.21мм2.

    Постройте доверительный интервал для математического ожидания, если 𝛼=0.05.
"""
X_, sigma, n, u = 10.3, 1.21, 16, 1.96

left_bound = X_ - 1.96 * sigma/math.sqrt(n)
right_bound = X_ + 1.96 * sigma/math.sqrt(n)

print("Step 9\n", "Bounds:", "%.2f" % left_bound, "< a <", "%.2f" % right_bound)

# Step 10
"""
    Постройте 95%-доверительный интервал для дисперсии 𝜎2 ошибок измерений, если  
    предполагается, что ошибки измерений подчиняются нормальному закону с мат.ожиданием 𝑎=0 и 
    неизвестной дисперсией 𝜎2.

    Формат ответа: границы интервала через запятую с пробелом, вводятся с точностью 3 
    знака после запятой. Сначала левая граница, потом правая

    Рассмотрите 𝑌=𝑛𝑆̃ 2𝜎2, где 𝑆̃ 2=1𝑛∑𝑛𝑖=1(𝑋𝑖−𝑎)2
    С.в. 𝑌 подчиняется распределению 𝜒2(𝑛) 
    Для нахождения квантилей распределения 𝜒2(𝑛) можно воспользоваться функцией qchisq() в R
"""
sample = np.loadtxt(os.path.join(root, "04_10.txt"), delimiter=",", unpack=False)

n = len(sample)
s = sum(sample**2)

left_bound = s / scp.chi2.ppf(0.975, df=n)
right_bound = s / scp.chi2.ppf(0.025, df=n)

print("Step 10\n", "Bounds:", "%.3f" % left_bound, "< simga2 <", "%.3f" % right_bound)

# Step 16
"""
Большая партия изделий может содержать некоторую долю дефектных. Поставщик утверждает, что эта 
доля составляет 5%, покупатель  предполагает, что доля дефектных изделий равна 10%.

По предложенной выборке постройте 95%-й приближенный доверительный интервал  для доли брака
в партии  и ответьте на вопрос: с большей степенью уверенности можно утверждать, что доля брака  
в партии больше 5%? (G - доля брака больше 5%, L - в противном случае)

Формат ответа: через запятую с пробелом

𝑆−,𝑆+, R

границы интервала 𝑆−,𝑆+ с точностью до 4го знака, R  = L или G
"""

sample = np.loadtxt(os.path.join(root, '04_16.txt'), delimiter=',', unpack=False)

n, m, u = len(sample), len(np.where(sample == 1)[0]), 1.96

S_minus = m/n - 1.96*math.sqrt(m/n * (1 - m/n)) / math.sqrt(n)
S_plus = m/n + 1.96*math.sqrt(m/n * (1 - m/n)) / math.sqrt(n)

print("Step 16\n", "Bounds:", "%.4f" % S_minus, "< p <", "%.4f" % S_plus)
```
